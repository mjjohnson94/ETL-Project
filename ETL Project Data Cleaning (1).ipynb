```python
import pandas as pd
```

```python
csv_file = 'Desktop/temp.csv'
```

```python
df = pd.read_csv(csv_file, encoding='cp1252')
```

```python
df.drop(df.columns[7:36], axis=1, inplace=True)
```

```python
del df['Months Code']
```

```python
del df['Element Code']
```

```python
del df['Y2015']
```

```python
del df['Y2016']
```

```python
del df['Y2017']
```

```python
del df['Y2018']
```

```python
del df['Y2019']
```

```python
df
```

```python
temp_filter_df = df.loc[(df['Area Code'] < 5000) & (df['Months'] == 'Meteorological year') &
                          (df['Element'] == 'Temperature change'), :]
```

```python
temp_filter_df
```

```python
del temp_filter_df['Area Code']
```

```python
del temp_filter_df['Months']
```

```python
temp_flip_df = temp_filter_df.melt(id_vars=["Area", "Element", "Unit"], var_name="Year", value_name="Temp")
```

```python
temp_flip_df = temp_flip_df.rename(columns={'Area':'Country'})
```

```python
temp_flip_df
```

```python
temp_flip_df['Year'] = temp_flip_df['Year'].str.strip('Y')
```

```python
clean_data = temp_flip_df.dropna()
clean_data
```

```python
csv_file2 = 'Desktop/greenhouse_gasses.csv'
```

```python
df2 = pd.read_csv(csv_file2)
```

```python
emission_filter_df = df2[['country_or_area', 'year', 'value']]
```

```python
emission_filter_df = emission_filter_df.rename(columns={'country_or_area':'Country', 'year': 'Year', 'value': 'Co2 emissions'})
emission_filter_df['Co2 emissions'] = emission_filter_df['Co2 emissions'].round(3)
emission_filter_df
```

```python
emission_filter_df['Year'] = emission_filter_df['Year'].astype(str)
```

```python
merge = pd.merge(clean_data, emission_filter_df, on=['Country', 'Year'])
merge
```
